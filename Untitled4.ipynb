```python
import matplotlib.pyplot as plt
import pandas as pd
mtcars=pd.read_csv('/content/shreee.csv')
trans_counts=mtcars['am'].value_counts()
trans_counts.plot(kind='bar')
plt.title('transmission of frequency')
plt.xlabel("transmission'(0=auto,1=manual)")
plt.ylabel("frequency")
plt.grid(True)
plt.show()



```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
mtcars=pd.read_csv('/content/shreee.csv')
sns.scatterplot(x='wt',y='mpg',data=mtcars)
plt.title('weight vs mpg')
plt.xlabel('weight')
plt.ylabel('mpg')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
mtcars=pd.read_csv('/content/shreee.csv')
sns.boxplot(y=mtcars['mpg'],color='green')
plt.title('distibution of the mpg ')
plt.ylabel('mpg')
plt.show()


```

```python
import matplotlib.pyplot as plt
import pandas as pd
mtcars=pd.read_csv('/content/shreee.csv')
plt.hist(mtcars['mpg'],bins=10,color='lightgreen',edgecolor='black')
plt.xlabel('mpg')
plt.ylabel('frequency')
plt.title('distibution')
plt.show()
```
